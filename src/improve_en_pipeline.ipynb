```python
import pandas as pd
import duckdb
```

```python
df = pd.read_csv("../data/labelled_feedback/2025-03-10_1209PM_SG_labelled_feedback_data.csv", encoding="utf-8")
df
```

```python
query = \
'''WITH cte AS (
    SELECT "Article ID", COUNT(*) AS feedbacks_per_article
    FROM df
    GROUP BY "Article ID"
    ORDER BY "Article ID"
)
SELECT * 
FROM df
INNER JOIN cte ON cte."Article ID" = df."Article ID"
order by feedbacks_per_article desc
;
'''

result = duckdb.query(query).df()
result.drop("Article ID_1", axis = 1, inplace=True)
result
# cte."Article ID", cte."Feedback 1", cte."Comments", cte."Label(s)", cte."Link to Article", cte."Type", df.feedbacks_per_article
```

```python
result.to_csv("../data/SG_final.csv", encoding="utf-8")
```
